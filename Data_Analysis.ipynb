```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
data_df =pd.read_csv('cleaned_data.csv')
data_df = data_df.drop(['Unnamed: 0'], axis=1)
data_df.head(3)
```

```python
dropped=data_df[data_df['Announce_to_Burial']<0]
dropped_=dropped.index
data_df=data_df.drop(dropped_)

dropped=data_df[data_df['Announce_to_Burial']>90]
dropped_=dropped.index
data_df=data_df.drop(dropped_)
```

```python
data_df.reset_index(drop=True, inplace=True)
data_df.Married.unique()
```

```python
pal = {1:"green", 0:"Pink"}
plt.subplots(figsize = (10,5))
ax = sns.barplot(x = "Gender", 
            y = "Fundraising", 
            data=data_df, 
            palette = pal,
            linewidth=2 )
plt.title("Fundraising/No Fundraising Gender Distribution", fontsize = 15)
plt.ylabel("% of Fundraising", fontsize = 15)
plt.xlabel("Gender",fontsize = 15);
plt.show()
```

```python
pal = {1:"seagreen", 0:"gray"}
sns.set(style="darkgrid")
plt.subplots(figsize = (15,8))
ax = sns.countplot(x = "Gender", 
                   hue="Fundraising",
                   data = data_df, 
                   linewidth=2, 
                   palette = pal
)



## Fixing title, xlabel and ylabel
plt.title("Gender Distribution - Fundaising vs No Fundaising", fontsize = 25)
plt.xlabel("Sex", fontsize = 15);
plt.ylabel("# ", fontsize = 15)

## Fixing xticks
#labels = ['Female', 'Male']
#plt.xticks(sorted(train.Sex.unique()), labels)

## Fixing legends
leg = ax.get_legend()
leg.set_title("Fundraised")
legs = leg.texts
legs[0].set_text("No")
legs[1].set_text("Yes")
plt.show()
```

```python
pal = {1:"green", 0:"Pink"}
plt.subplots(figsize = (10,5))
ax = sns.barplot(x = "Married", 
            y = "Fundraising", 
            data=data_df, 
            palette = pal,
            linewidth=2 )
plt.title("Fundraising/No Fundraising Married Distribution", fontsize = 15)
plt.ylabel("% of Fundraising", fontsize = 15)
plt.xlabel("Married",fontsize = 15);
plt.show()
```

```python
pal = {1:"green", 0:"Pink"}
plt.subplots(figsize = (10,5))
ax = sns.barplot(x = "Color", 
            y = "Fundraising", 
            data=data_df, 
            palette = pal,
            linewidth=2 )
plt.title("Fundraising/No Fundraising Color Distribution", fontsize = 15)
plt.ylabel("% of Fundraising", fontsize = 15)
plt.xlabel("Color",fontsize = 15);
plt.show()
```

```python
pal = {0:"green", 1:"Pink",2:"blue", 3:"red",4:"purple", 5:"grey",6:"black"}
plt.subplots(figsize = (10,5))
ax = sns.barplot(x = "Burial_Day", 
            y = "Fundraising", 
            data=data_df, 
            palette = pal,
            linewidth=2 )
plt.title("Fundraising/No Fundraising Burial_Day Distribution", fontsize = 15)
plt.ylabel("% of Fundraising", fontsize = 15)
plt.xlabel("Burial_Day",fontsize = 15);
plt.show()
```

```python
# Kernel Density Plot
fig = plt.figure(figsize=(10,5),)
ax=sns.kdeplot(data_df.loc[(data_df['Fundraising'] == 0),'Size'] , color='gray',shade=True,label='no Fundraising')
ax=sns.kdeplot(data_df.loc[(data_df['Fundraising'] == 1),'Size'] , color='g',shade=True, label='fundraising')
plt.title('Size Distribution - Fundraising V.S. No Fundraising', fontsize = 15)
plt.xlabel("Size", fontsize = 15)
plt.ylabel('Frequency', fontsize = 15);
plt.show()
```

```python
fig = plt.figure(figsize=(10,5),)
ax=sns.kdeplot(data_df.loc[(data_df['Fundraising'] == 0),'No_of_Relatives'] , color='gray',shade=True,label='no Fundraising')
ax=sns.kdeplot(data_df.loc[(data_df['Fundraising'] == 1),'No_of_Relatives'] , color='g',shade=True, label='Fundraising')
plt.title('No_of_Relatives Distribution - Fundraising V.S. No Fundraising', fontsize = 15)
plt.xlabel("Number of relatives", fontsize = 15)
plt.ylabel('Frequency', fontsize = 15);
plt.show()
```
